```python
import pandas as pd

la_apts = pd.read_csv ('Resources/la_apts_coords.csv')
coords_df = la_apts[['Latitude', 'Longitude']]
coords_df
```

```python
import googlemaps
gmaps = googlemaps.Client(key='AIzaSyCnYHZHMZCq0TT1eaV_tXlhpbPZ6tl-xkU')

iterations = 0

zipcodes = []

for row in coords_df[['Latitude', 'Longitude']].itertuples():
    
    reverse_geocode_result = gmaps.reverse_geocode((row[1], row[2]))
        
    for component in reverse_geocode_result[0]['address_components']:
        zipcode = None
                
        if "postal_code" in component["types"]:
            zipcode = component['long_name']
            break
            
    zipcodes.append(zipcode)

    iterations += 1
    print(str(iterations) + " zips converted successfully!")

print("\n")

print("zip conversion complete!")
```

```python
la_apts['Zipcode'] = zipcodes
la_apts
```

```python
import googlemaps
gmaps = googlemaps.Client(key='AIzaSyCnYHZHMZCq0TT1eaV_tXlhpbPZ6tl-xkU')

iterations = 0

neighborhoods = []

for row in coords_df[['Latitude', 'Longitude']].itertuples():
    
    reverse_geocode_result = gmaps.reverse_geocode((row[1], row[2]))
        
    for component in reverse_geocode_result[0]['address_components']:
        neighborhood = None
                
        if "neighborhood" in component["types"]:
            neighborhood = component['long_name']
            break
            
    neighborhoods.append(neighborhood)

    iterations += 1
    print(str(iterations) + " neighborhoods converted successfully!")

print("\n")

print("neighborhood conversion complete!")
```

```python
la_apts['Neighborhood'] = neighborhoods

la_apts_clean = la_apts.drop(columns=['neighborhood'])
```

```python
la_apts_clean.to_csv('Resources/la_apts_clean.csv')
```
